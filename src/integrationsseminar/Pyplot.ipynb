```python
#Einfacher Plot
import matplotlib.pyplot as plt
plt.plot([1,5,3])
plt.show()
```

```python
#Viele verschiedene Parameter
import numpy as np
xPoints = [1,2,3,5]
yPoints = [3,8,1,10]
#plt.plot(xPoints, yPoints, marker = 'o',linestyle = 'dotted', color='red')

plt.plot(xPoints, yPoints, marker = 'o',ms=20,mfc='b', ls = '-', c='r',lw=10) #Kurzschreibweise
yPoints2 = np.array([3,4,7])
plt.plot(yPoints2)
plt.title('Basic plot', loc = 'right')
font1 = {'family':'serif','color':'blue','size':20}
plt.xlabel("Zeit in Sekunden", font1)
plt.grid()

plt.show()
```

```python
#Mehrere Plots
plt.subplot(1, 2, 1)
plt.plot(xPoints,yPoints)
plt.title("Arbeiten")
plt.subplot(1, 2, 2, facecolor='red')
plt.title("Test") #Hier sieht man auch dass es implizit auf den letzten Plot geht
plt.plot([5,4,1])
plt.suptitle("Lohnt sich das?")
plt.show()
```

```python
#Explizite Aufrufe – Objekt orientiert
fig = plt.figure(figsize=(12,8))
axes = fig.subplots(1,2)
(ax1,ax2) = axes
        #axes[1].plot(xPoints,yPoints)
ax1.plot(xPoints,yPoints)
fig.set_facecolor('red')
ax2.set_title('rechts')
ax1.set_facecolor('blue')
plt.show()
```
