```python
#Einfacher Plot â€“ selbst coden
import matplotlib.pyplot as plt
plt.plot([1,5,3])
plt.show()
```

```python
#Viele verschiedene Parameter
import numpy as np
yPoints2 = np.array([3,4,7])
xPoints = [1,2,3,5]
yPoints = [3,8,1,10]
plt.plot(xPoints, yPoints, marker = 'o',linestyle = 'dotted', color='red')
#plt.plot(xPoints, yPoints, marker = 'o',ms=20,mfc='b', ls = '-', c='r',lw=10) #Kurzschreibweise
plt.plot(yPoints2)
plt.title('Basic plot', loc = 'right')
font1 = {'family':'serif','color':'blue','size':20}
plt.xlabel("Zeit in Sekunden", font1)
plt.grid()
plt.show()
```

```python
#Mehrere Plots
plt.subplot(1, 2, 1)
plt.plot(xPoints,yPoints)
plt.title("Arbeiten")
plt.subplot(1, 2, 2, facecolor='red')
plt.title("Test") #Hier sieht man auch dass es implizit auf den letzten Plot geht
plt.title("Aktien")
plt.plot([5,4,1])
plt.suptitle("Lohnt sich das?")
plt.show()
```

```python
#Explizite Aufrufe â€“ Objekt orientiert
#fig, (ax1,ax2) = plt.subplots(1,2)
fig = plt.figure(figsize=(12,8))
axes = fig.subplots(1,2)
(ax1,ax2) = axes
#axes[1].plot(xPoints,yPoints)
ax1.plot(xPoints,yPoints)
fig.set_facecolor('red')
fig.suptitle('Figure')
ax2.set_title('rechts')
ax1.set_facecolor('blue')
plt.show()
```

```python
#Scatter
plt.scatter(xPoints,yPoints)
x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(x, y, color='red', alpha=0.4)
plt.plot([50,35,40])
plt.show()
```

```python
#Bar
xLabels = ['A','B','C','D']
plt.bar(xLabels,yPoints, width=0.2)
plt.show()
```

```python
#Pie
myexplode = [0.3, 0, 0, 0]
plt.pie([2,5,3,10], labels=xLabels, startangle = 90, explode=myexplode)
plt.legend(title = "Buchstaben")
plt.show()
#Startet bei der X-Achse und zeichnet gegen den Uhrzeigersinn
```
